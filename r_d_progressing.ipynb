````python
with open('observe/readme_hand.csv','r',encoding='utf-8') as f:
    readme_csv = f.readlines()
print("文件加载完毕")
````

````python
# 去除 <sourve ...> 标签
readme_list = readme_csv[1:]
readme_csv_len = len(readme_list)
print("仓库大小为：",readme_csv_len)
readme_source = []
source_count = 0
for i in range(readme_csv_len):
    one_readme = readme_list[i]
    one_readme_len = len(one_readme)
    #print(one_readme_len)
    new_readme = ''
    start = 0
    j = 0
    while j <= one_readme_len:
        if one_readme[j:(j+8)]=='<source ':
            source_start = j
            source_count+=1
            print("仓库",i,"存在",one_readme[j:(j+30)],"...标签待去除")
            new_readme+=one_readme[start:j]
            new_readme+=' '
            while one_readme[j]!='>':
                j+=1
            start = j+1
            print("去掉的标签为：",one_readme[source_start:start],"，本标签长度为：",start-source_start)
        else:
            j += 1
    new_readme += one_readme[start:j]
    readme_source.append(new_readme)
    #print(len(new_readme))   
print(source_count)
print(len(readme_source))
````

````python
# 存储去除 <sourve ...> 标签结果
csv_filename = 'observe/readme_source.csv'
with open(csv_filename, 'w', encoding='utf-8') as f:
    f.write('readme\n')
    for i in range(len(readme_source)):
        f.write(str(readme_source[i])) 
print(f"Data saved to {csv_filename} successfully.")
````
